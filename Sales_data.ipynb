# Q.2 What is the best year according to sales?

```python
yearly_sales = df.groupby(["YEAR_ID"])["SALES"].sum().reset_index()
yearly_sales
```

```python
# set the figure size
plt.figure(figsize=(7,5))

plt.title("YEARLY SALES")
plt.xlabel("YEARS")
plt.ylabel("SALES ( IN MILLIONS)")

z = sns.barplot(x="YEAR_ID", y="SALES", data=yearly_sales, hue="YEAR_ID", palette="mako", edgecolor="black", width=0.3)
sns.set_style("darkgrid")

# to add lables to each bar
tb = ["3.51M", "4.72M", '1.79M']

# adding to each container
z.bar_label(z.containers[0], labels=tb, padding=3)
z.bar_label(z.containers[1], labels=tb, padding=3)
z.bar_label(z.containers[2], labels=tb, padding=3)

# adujust the legend size
plt.legend(prop={"size":12})
plt.plot()
```

#### Observations :

1) In 2003, Sales were around 3.5 Million
2) In 2004, Sales were around 4.7 Million,Highest sales in the 2004 Year.
3) In 2005, Sales were around 1.8 Million in just 5 Month, which was the very progressive year.

****Barplot represent the progressive growth of Sales****

# Q3: Which quarter is the best for each product line?

```python
quarter = df.groupby(["PRODUCTLINE"])["QTR_ID"].value_counts().reset_index(name='COUNTS')
quarter
```

```python
# Set the width and height of the figure
plt.figure(figsize=(9,6))

kwargs = {"edgecolor":"black", "linewidth":1}

sns.stripplot(data = quarter, x="PRODUCTLINE", y="COUNTS", hue="QTR_ID", palette='bright', size=15, **kwargs)
sns.set_style("whitegrid")

plt.title("PRODUCTS SOLD ACCORDING TO QUARTER")
plt.xlabel("PRODUCTS")
plt.ylabel("NUMBER OF ITEMS SOLD")
plt.show()
```

#### Observation 

- In 4th quarter of each products had maximum sales.
- In product categories, "Classic Cars" & "Vintage Cars" are the most demanded vehicles.
- The reason may be arrival of the Christmas Holidays.

# Q4: What product sold the most? Why do you think it sold the most?

```python
products = df.groupby(["PRODUCTLINE"])["SALES"].sum().reset_index()
products
```

```python
products = df.groupby(["PRODUCTLINE"])['SALES'].sum().reset_index()
prices = df.groupby(["PRODUCTLINE"])["MSRP"].min().reset_index()  # MSRP - manufactured suggested retail price

result = pd.merge(products, prices)
result.SALES/100000
```

```python
# set the width and height of the figure
plt.figure(figsize=(10,6))


ax= sns.barplot(data=result, x ="PRODUCTLINE", y = "SALES", hue="PRODUCTLINE", palette="inferno",
                             edgecolor="black", dodge=False, width=0.4)
ax2 = plt.twinx()
sns.lineplot(data=result, x="PRODUCTLINE", y='MSRP',ax=ax2, errorbar=None, color="b")

# adding labels
tb = ["3.91M", "1.16M", '976K', '714K', '226K', '1.12M', '1.90M' ]

# ADD TO THE CONTAINER
ax.bar_label(ax.containers[0], labels=tb, padding=3)
ax.bar_label(ax.containers[1], labels=tb, padding=3)
ax.bar_label(ax.containers[2], labels=tb, padding=3)
ax.bar_label(ax.containers[3], labels=tb, padding=3)
ax.bar_label(ax.containers[4], labels=tb, padding=3)
ax.bar_label(ax.containers[5], labels=tb, padding=3)
ax.bar_label(ax.containers[6], labels=tb, padding=3)

# adding title to the plot
plt.title("PRODUCTS SOLD vs SALES / MSRP")
ax.set(xlabel="PRODUCTS")
plt.show()
```

#### Observation 

- The above graph shows that the products with the lowest price were sold the most, whereas the products with a high price were the least popular.

- From the plot, 'Classic Cars" & "Vintage Cars" are the most demanded products, 'Trains" was the least demanded product.

# Q5: Who was the top customer?

```python
top_customers = df.groupby(["CUSTOMERNAME"]).sum().sort_values("SALES", ascending=False).head(5)
top_customers = top_customers[['SALES']].round(3)
top_customers.reset_index(inplace=True)
top_customers.head()
```

```python
top_customers.SALES/1000
```

```python
plt.figure(figsize= ( 8,6))


# main plot
z = sns.barplot(data= top_customers, x = "CUSTOMERNAME", y="SALES", hue='CUSTOMERNAME', palette="icefire",
                                     edgecolor="black", dodge=False, width=0.3)
sns.set_style("whitegrid")

# adding title 
plt.title('TOP 5 CUSTOMERS')
z.set(xlabel="CUSTOMERS", ylabel='SALES')

# tb
tb = ["912K", '654K', "200K", "197K", "180K"]
z.bar_label(z.containers[0], labels=tb, padding=2)
z.bar_label(z.containers[1], labels=tb, padding=2)
z.bar_label(z.containers[2], labels=tb, padding=2)
z.bar_label(z.containers[3], labels=tb, padding=2)
z.bar_label(z.containers[4], labels=tb, padding=2)

plt.xticks(rotation=70)
z.set_yticklabels(["0", "200K", "400K", "600K", "800K"])

# ADJUST THE LEGEND
plt.legend(prop={"size":10})
plt.legend(loc="upper right")
plt.show()
```

#### Observations

- "Euro Shopping Channel" was the top customer buying the most products.

# Q6: Which country has the best sales?

```python
country = df.groupby(["COUNTRY"])['SALES'].sum().reset_index()

# short form of country
short_form = ['AUS','AUT','BEL','CAN','DNK','FIN','FRA','DEU','IRL','ITA','JPN','NOR','PHL','SGP','ESP','SWE','CHE','GBR','USA']
country["iso_alpha"] = short_form
country
```

```python
plt.figure(figsize=(16,16))

# choropleth map using plotly
fig = px.choropleth(country, locations = "iso_alpha",color= "SALES", scope="world", hover_name="COUNTRY")

# UPDATING THE MAP by adding the title
fig.update_layout(title= dict( 

                              text="Countries according to their Sales",
                              xanchor="center",
                              x=0.5,
                              yref="paper",
                              y=1,
                              pad={"b":60}
                 )
)

fig.show()
```

#### Observation

- "USA" had the best sales in the world.
-  After that "Spain" & "France" was followed by "USA" 

# Q7. Which city has the best sales? Why?

```python
cities = df.groupby([ "CITY"])['SALES'].sum().reset_index()
cities.columns = ['city', "sales"]
cities.sort_values(by="sales", ascending=False)
```

```python
# adding cities data for latitude and longitude
cities2 = pd.read_csv("/kaggle/input/world-cities/worldcities.csv", encoding=('ISO-8859-1'), index_col=0)

# # extract required column 
# city_df = cities2[["iso3", "lat", "lng"]]
# city_df
cities2
```

```python
# merge the both dataframe cities and city_df
ct_df = pd.merge(cities, cities2, on="city", how="inner")
ct_df = ct_df[["city", "sales", "lat", "lng", "iso3"]]
ct_df
```

```python
plt.figure(figsize = (16,16))

fig = px.scatter_geo(ct_df, lat="lat", lon="lng", color="iso3", hover_name="city", size="sales")

# adding title and centering it
fig.update_layout(title= dict( text= "Cities according to their sales", xanchor="center", x=0.5, yref="paper", yanchor="bottom", y=0.9, pad={"b":60}))
fig.show()
```

#### Observations

- City of "Madrid" had the best sales in the world followed by "San Rafael" and "New York City".
